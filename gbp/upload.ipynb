```python
g = data['gbp_gradient_binary'].as_matrix()
c = data['correct'].as_matrix()
```

```python
usdx_gradient = data['gbp_gradient'].as_matrix()
temp = np.zeros(6348)
index = np.logical_and(g, c)
temp[index] = usdx_gradient[index]
data['gbp_gradient_up_correct'] = temp
data.head()
```

```python
usdx_gradient = data['gbp_gradient'].as_matrix()
temp = np.zeros(6348)
index = np.logical_and(np.logical_not(g), c)
temp[index] = usdx_gradient[index]
data['gbp_gradient_down_correct'] = temp
data.head()
```
